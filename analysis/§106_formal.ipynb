# Clause Dependency


> In the datasets made with Text-Fabric this variable is called “main_sub” and has the following values: “Main” (main clause), “SubArg” (subordinate, argument clause), “SubAdv” (subordinate, adverbial clause), “SubMod” (subordinate, attributive clause).
The difference between adverbial clauses and argument clauses is that the latter type is a necessary part of a clause whereas the adverbial clauses can be omitted without making a clause ungrammatical. There is actually a wide variety of subordinate clauses which fall under the category of adverbial clauses in our research, and linguists differ in how they describe the adverbial clause. (Naaijer 2020:39)

See also from the tagger:

```
    if ccr in {'Subj', 'Objc', 'Cmpl', 'PreC', 'Voct', 'Frnt'}:
        in_dep_r += 'SubArg'
        
    elif ccr in {'Attr', 'RgRc', 'Spec'}:
        in_dep_r += 'SubMod'
        
    elif ccr in {'Adju', 'PrAd'}:
        in_dep_r += 'SubAdv'
```

# Independent Clause Types

## Mother Clause Types

NB: when the following restrictions apply:

1. mother clause is headed by indicative verb
2. mother clause is within the same domain
3. this clause is a main clause, not subordinate or adverbial
4. clause domain is Q or N, not D or ?

````python
av_df.columns
````

````python
mo_vtypes = ['qtl', 'wqtl', 'yqtl', 'ptcp', 'wayq']
mother_df = av_df[
    (av_df.mother_verbtype.isin(mo_vtypes)) # restrict to indicative verbs
    & ((av_df.mother_domain2) == (av_df.domain2)) # agreement in domain
    & (av_df.clause_rela == 'Main') # retrict to main clauses
    & (av_df.domain2.isin(['Q', 'N']))
]
av_mother = PivotProp(mother_df, ['verb_form', 'domain2'], 'mother_verbtype')

av_mother.ct
````

## Daughter Clause Types

NB: when the following restrictions apply:

1. mother clause is headed by indicative verb
2. daughter clause is in same domain
2. daughter clause is a main clause, not subordinate or adverbial
3. clause domain is Q or N, not D or ?

````python
daught_df = av_df[
    (av_df.daught_verbtype.isin(mo_vtypes)) # restrict to indicative verbs
    & ((av_df.mother_domain2) == (av_df.domain2)) # agreement in domain
    & (av_df.daught_rela == 'Main') # retrict to main clauses
    & (av_df.domain2.isin(['Q', 'N']))
]
av_daught = PivotProp(daught_df, ['verb_form', 'domain2'], 'daught_verbtype')

av_daught.ct
````
