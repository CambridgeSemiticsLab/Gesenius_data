## Book

```python
av_book = PivotProp(av_df, 'verb_form', 'book')

av_book.ct
```

```python
av_book.pr2
```

```python
av_book.pr2['qtl'].sort_values(ascending=False).head()
```

```python
av_book_fish, avb_odds = my_stats.apply_fishers(av_book.ct, 0, 1)

av_book_fish
```

This is a lot of values. We want to see if any particular book is especially associated with qatal. I expect, ft. Qohelet may be so.
I'll use the Fisher's test for this.

```python
av_book_fish, avb_odds = my_stats.apply_fishers(av_book.ct, 0, 1)

av_book_fish
```

Look for strong responses with qtl.

```python
avb_fish_qtl = av_book_fish.loc['qtl'].sort_values(ascending=False)

avb_fish_qtl.head(15)
```

Fisher's test appears to miss cases like Ecclesiastes, which have a high share of qtl compared to qtl's usual share elsewhere. 

Let's use a unidirectional test, deltaP, instead. We'll use the books as the cue and verbform as the conditional response.

```python
av_book_dp = my_stats.apply_deltaP(av_book.ct.T, 0, 1)

av_book_dp.head()
```

Look for strong cases with qtl.

```python
av_book_dp['qtl'].sort_values(ascending=False).head(10)
```

Ecclesiastes still doesn't show!

```python
av_book_qtl = av_book.ct[avb_fish_qtl[avb_fish_qtl > 10].index].T

avb_qtl_pr = prop_table(av_book_qtl)

av_book_qtl
```

```python
avb_qtl_pr
```

## Mother Clause Types

NB: when the following restrictions apply:

1. mother clause is headed by indicative verb
2. mother clause is within the same domain
3. this clause is a main clause, not subordinate or adverbial
4. clause domain is Q or N, not D or ?

```python
av_df.columns
```

```python
mo_vtypes = ['qtl', 'wqtl', 'yqtl', 'ptcp', 'wayq']
mother_df = av_df[
    (av_df.mother_verbtype.isin(mo_vtypes)) # restrict to indicative verbs
    & ((av_df.mother_domain2) == (av_df.domain2)) # agreement in domain
    & (av_df.clause_rela == 'Main') # retrict to main clauses
    & (av_df.domain2.isin(['Q', 'N']))
]
av_mother = PivotProp(mother_df, ['verb_form', 'domain2'], 'mother_verbtype')

av_mother.ct
```

```python
av_mother.pr
```

```python
av_mother.oneN
```

```python
show_text(
    mother_df[
        (mother_df.verb_form == 'qtl')
        & (mother_df.mother_verbtype == 'wayq')
    ]
, spread=10, col_default=['ref', 'text_full', 'sentence', 'mother_clause'])
```

## Daughter Clause Types

NB: when the following restrictions apply:

1. mother clause is headed by indicative verb
2. daughter clause is in same domain
2. daughter clause is a main clause, not subordinate or adverbial
3. clause domain is Q or N, not D or ?

```python
daught_df = av_df[
    (av_df.daught_verbtype.isin(mo_vtypes)) # restrict to indicative verbs
    & ((av_df.mother_domain2) == (av_df.domain2)) # agreement in domain
    & (av_df.daught_rela == 'Main') # retrict to main clauses
    & (av_df.domain2.isin(['Q', 'N']))
]
av_daught = PivotProp(daught_df, ['verb_form', 'domain2'], 'daught_verbtype')

av_daught.ct
```
