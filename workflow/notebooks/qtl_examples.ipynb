```python
def make_text_examples(df, suffix=''):
    """Build copy and pastable text samples."""
    
    exs = []
    
    for node in df.index:
        ref = df.loc[node]['ref']
        esv = df.loc[node][f'esv{suffix}']
        niv = df.loc[node][f'niv{suffix}']
        bhs = df.loc[node]['clause_atom']
            
        if niv == esv:
            ex = f'{esv} (ESV, NIV | BHS {bhs} {ref})'
        else:
            ex = f'{esv}, {niv} (ESV, NIV | BHS {bhs} {ref})'
        
        exs.append(ex)
        
    return exs
```
